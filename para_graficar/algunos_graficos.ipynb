```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sy
from para_graficar import ejes, graficar

x = sy.symbols('x')
f = sy.sqrt(x) - sy.cos(x)
fig, ax = plt.subplots()
ejes()
graficar(f, [0,1, 3], color='red')
ax.set_title(r"$\sqrt{x}-\cos (x)$")
plt.show()
```
