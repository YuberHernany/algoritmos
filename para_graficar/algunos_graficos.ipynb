```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-6,6)
mu = 0
# sigma = 1
sigmas = [1/4, 1/2, 1, 2, 4]
fig, ax = plt.subplots()
ax.spines['bottom'].set_position('zero')
ax.spines['left'].set_position('zero')
for sigma in sigmas:
    cte = 1/(np.sqrt(2 * np.pi) * sigma)
    y = cte * np.exp(-(x-mu)**2/(2 * sigma**2))

    ax.plot(x,y)
plt.show()
```
